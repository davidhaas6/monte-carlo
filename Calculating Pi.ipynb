# Calculating the value of Pi using Monte Carlo approximation

```python
import matplotlib.pyplot as plt
import numpy as np

n_samples = 10000
center = (.5,.5)
radius = .5
```

```python
# Sample randomly, determine if they're in the circle
samples = np.random.rand(n_samples*2).reshape(-1,2)  # rows of xy pairs
dists = np.linalg.norm(center - samples, axis=1)
in_circle = dists < radius

# Plotting
plt.scatter(samples[:,0], samples[:,1], s=.1, c=in_circle)
plt.title("Random Samples")
plt.gcf().set_size_inches(5,5)
```

# Estimate the value of Pi

```python
area = sum(in_circle) / n_samples
pi = area / (radius**2)
print(" ",pi)
```
